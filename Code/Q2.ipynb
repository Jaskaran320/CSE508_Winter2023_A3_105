```python
import matplotlib.pyplot as plt
import networkx as nx

PATH = '../Dataset/p2p-Gnutella08.txt'
```

```python
G = nx.read_edgelist(PATH, create_using=nx.DiGraph(), delimiter='\t')
pagerank_scores = nx.pagerank(G)
hubs, authorities = nx.hits(G)
```

```python
print('PageRank Scores:')
for node, score in pagerank_scores.items():
    print(f'Node {node}: {score:.6f}')
```

```python
print('Hubs Scores:')
for node, score in hubs.items():
    print(f'Node {node}: {score:.6f}')
```

```python
print('Authorities Scores:')
for node, score in authorities.items():
    print(f'Node {node}: {score:.6f}')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
ax1.scatter(list(pagerank_scores.values()), list(authorities.values()))
ax1.set_xlabel('PageRank Score')
ax1.set_ylabel('Authority Score')
ax1.set_title('PageRank vs Authority Scores')
ax2.scatter(list(pagerank_scores.values()), list(hubs.values()))
ax2.set_xlabel('PageRank Score')
ax2.set_ylabel('Hub Score')
ax2.set_title('PageRank vs Hub Scores')
plt.show()
```
